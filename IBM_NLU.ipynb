```python
import json
from ibm_watson import NaturalLanguageUnderstandingV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
from ibm_watson.natural_language_understanding_v1 import Features, \
                                                        CategoriesOptions,\
                                                        ClassificationsOptions,SentimentOptions,\
                                                        EntitiesOptions

authenticator = IAMAuthenticator('your api key')
natural_language_understanding = NaturalLanguageUnderstandingV1(
    version='2022-04-07',
    authenticator=authenticator
)

natural_language_understanding.set_service_url('your ibm url')
```

<div style="background-color:white; color:black; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;">
Ensuite on va jouer avec le text en explorant les fonctionallités offert par l'api:
    
</div>

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">
Categories feature request 
</div>

```python
response = natural_language_understanding.analyze(
    url='https://www.chess.com/fr/article/view/carlsen-fischer-et-capablanca-avaient-ils-tort',
    features=Features(categories=CategoriesOptions(limit=3))).get_result()

print(json.dumps(response, indent=2))
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">
Classifications feature request
</div>

```python
text = ''' Le ciel était sombre et chargé de nuages gris, reflétant mon état d'âme morose. Les gouttes de pluie tombaient lentement,
comme les larmes silencieuses d'un cœur brisé. Les rues étaient désertes, tout comme mon esprit, empli d'une tristesse profonde.
Chaque pas était lourd, portant le fardeau des souvenirs douloureux. Les rires joyeux des jours passés semblaient maintenant
lointains, perdus dans l'obscurité de la mélancolie. Les ombres de la solitude dansaient autour de moi, me rappelant cruellement
l'absence de chaleur humaine. Dans cet océan de tristesse, je cherchais désespérément une lueur d'espoir, mais même les étoiles
semblaient avoir perdu leur éclat.
'''
response = natural_language_understanding.analyze(
    text=text,
    features=Features(classifications=ClassificationsOptions(model='tone-classifications-en-v1'))
).get_result()
print(json.dumps(response, indent=2))
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">
Concepts feature request</div>

```python
Text_Concepts = '''
"In the quiet solitude of the moonlit night, I find myself pondering the complexities of life. 
The echoes of past memories resonate in the silence, each one carrying a story of joy or sorrow.
As I gaze at the stars above, I am reminded of the vastness of the universe and the fleeting nature of our existence.
The weight of unspoken words hangs in the air, and the shadows of unfulfilled dreams linger in the corners of my mind.
In this moment of reflection, the concept of time becomes both a companion and a betrayer, 
weaving a tapestry of experiences that define the human journey."
'''
response = natural_language_understanding.analyze(
    text=Text_Concepts,
    features=Features(concepts=ConceptsOptions(limit=3))
).get_result()
print(json.dumps(response, indent=2))
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">
Sentiment Analysis:</div>

```python
text_to_analyze = "I am so happy today! Everything is going well." 

response = natural_language_understanding.analyze(
    text=text_to_analyze,
    features=Features(sentiment=SentimentOptions())
).get_result()

print(json.dumps(response, indent=2))
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">
Entity Analysis:</div>

```python
text_to_analyze = "Apple Inc. is planning to launch a new product next month.IBM is good" 
response = natural_language_understanding.analyze(
    text=text_to_analyze,
    features=Features(entities=EntitiesOptions())
).get_result()
print(json.dumps(response, indent=2))
```

<div style="background-color:orangered; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;text-align:center">

<div style="color:brown; font-size:20px; font-family:Verdana; padding:10px;font-weight:bold;">
Fichier JSON:
</div>

```python
file="piaf-v1.1.json"
```

```python
with open('piaf-v1.1.json', 'r', encoding='utf-8') as file:
    dataset = json.load(file)
# Extract text data from the dataset
texts = [item['paragraphs'][0]['context'] for item in dataset['data']]
# Function to analyze text using Watson NLU
def analyze_text(text):
    response = natural_language_understanding.analyze(
        text=text,
        language='fr', 
        features=Features(
            sentiment=SentimentOptions(),
            entities=EntitiesOptions(),
            concepts=ConceptsOptions(limit=3)
        )
    ).get_result()
    return response
# Analyze each text in the dataset
for idx, text in enumerate(texts):
    print(f"\nAnalyzing Text {idx + 1}:\n{text}") 
    # Perform analysis
    analysis_result = analyze_text(text)   
    # Print sentiment analysis result
    sentiment_label = analysis_result['sentiment']['document']['label']
    print(f"Sentiment Analysis: {sentiment_label}")  
    # Print entity analysis result
    entities = analysis_result['entities']
    print("Entities:")
    for entity in entities:
        print(f"  - {entity['type']}: {entity['text']}")   
    # Print concept analysis result
    concepts = analysis_result['concepts']
    print("Concepts:")
    for concept in concepts:
        print(f"  - {concept['text']} ({concept['relevance']})")
```

<div style="color:brown; font-size:20px; font-family:Verdana; padding:10px;font-weight:bold;">
Fichier txt:
</div>

```python
# Read text from a file
def read_text_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# File path to your text file
file_path = 'Descriptif de piaf-v1.1.txt'

# Read text from the file
text_from_file = read_text_from_file(file_path)

# Perform analysis
analysis_result = analyze_text(text_from_file)

# Print sentiment analysis result
sentiment_label = analysis_result['sentiment']['document']['label']
print(f"Sentiment Analysis: {sentiment_label}")

# Print entity analysis result
entities = analysis_result['entities']
print("Entities:")
for entity in entities:
    print(f"  - {entity['type']}: {entity['text']}")

# Print concept analysis result
concepts = analysis_result['concepts']
print("Concepts:")
for concept in concepts:
    print(f"  - {concept['text']} ({concept['relevance']})")
```
